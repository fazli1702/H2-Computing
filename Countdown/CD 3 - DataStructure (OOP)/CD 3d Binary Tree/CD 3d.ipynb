```python
# Task 3.1
class Node:
    def __init__(self):
        self.data = ''
        self.leftptr = -1
        self.rightptr = -1
        
    def setData(self, s):
        self.data = s
        
    def setLeftPtr(self, x):
        self.leftptr = x
    
    def setRightPtr(self, y):
        self.rightptr = y
        
    def getData(self):
        return self.data
    
    def getLeftPtr(self):
        return self.leftptr
    
    def getRightPtr(self):
        return self.rightptr
    

class Tree:
    def __init__(self):
        self.root = -1
        self.thisTree = []
        self.NextFree = 0
        
    def add(self, newItem):
        newNode = Node()
        newNode.setData(newItem)
        self.thisTree.append(newNode)
        
        if self.root == -1:
            self.root = 0
            self.NextFree += 1
            
        else:
            currIndex = self.root
            while True:
                currNode = self.thisTree[currIndex]
                if newItem < currNode.getData():
                    if currNode.getLeftPtr() == -1:
                        currNode.setLeftPtr(self.NextFree)
                        self.NextFree += 1
                        break
                    else:
                        currIndex = currNode.getLeftPtr()
                        
                else:
                    if currNode.getRightPtr() == -1:
                        currNode.setRightPtr(self.NextFree)
                        self.NextFree += 1
                        break
                    else:
                        currIndex = currNode.getRightPtr()
            
    # Task 4.3
    def postOrderTraversal(self, currIndex=0):
        if self.root == -1:
            return 'Empty tree'
        else:
            # LRN
            currNode = self.thisTree[currIndex]
            if currNode.getLeftPtr() != -1:
                self.postOrderTraversal(currNode.getLeftPtr())
            if currNode.getRightPtr() != -1:
                self.postOrderTraversal(currNode.getRightPtr())
            print(currNode.getData())
                        
                        
    def print(self):
        print('root:', self.root)
        print('nextFree:', self.NextFree)
        print('%-15s %-15s %-15s %-15s' % ('Index', 'Data', 'leftptr', 'righptr'))
        for i, ele in enumerate(self.thisTree):
            print('%-15s %-15s %-15s %-15s' % (i, ele.getData(), ele.getLeftPtr(), ele.getRightPtr()))
```

```python
# Task 3.2
tree = Tree()
tree.add('Tiger')
tree.add('Lemur')
tree.add('Bat')
tree.add('Yak')
tree.add('Ostrich')
tree.add('Raccoon')
tree.add('Macaw')
tree.add('Zebra')
tree.print()
tree.postOrderTraversal()
```
