```python
# Task 3.1
class ListNode:
    def __init__(self):
        self.Name = ''
        self.Pointer = 0
    
    def SetName(self, Name):
        self.Name = Name
        
    def SetPointer(self, Pointer):
        self.Pointer = Pointer
        
    def GetName(self):
        return self.Name
    
    def GetPointer(self):
        return self.Pointer
    
    
class LinkedList:
    def __init__(self):
        self.Node = [ListNode() for i in range(11)]
        self.Start = 0
        self.NextFree = 1
        for i in range(10):
            if i == 0:
                self.Node[i] = None
            else:
                self.Node[i].SetPointer(i+1)
                
    def Insert(self, name, position):
        newNode = self.Node[self.NextFree]
        newNode.SetName(name)
        if position == 1:
            # empty list
            if self.Start == 0:
                newNode.SetPointer(0)
                self.Start = self.NextFree
                self.NextFree += 1
            # list already populated
            else:
                newNode.SetPointer(self.Start)
                self.Start = self.NextFree
                self.NextFree += 1
        
        else:
            # find node to link with newNode
            prevNode = self.Node[self.Start]
            for _ in range(position-2):
                prevNode = self.Node[prevNode.GetPointer()]
            i = self.NextFree
            self.NextFree = newNode.GetPointer()
            newNode.SetPointer(prevNode.GetPointer())
            prevNode.SetPointer(i)
            
                
                
    def Delete(self, position):
        if position == 1:
            delNode = self.Node[self.Start] #Node to be deleted
            delNode.SetName('')
            newStart = delNode.GetPointer() #New Start value
            newNextFree = self.Start  #New NextFree
            
            delNode.SetPointer(self.NextFree)
            self.Start = newStart
            self.NextFree = newNextFree
            
                
        else:
            pass
            
            
    
    def Length(self):
        if self.Start == 0:
            return 0
        else:
            i = 0
            currNode = self.Node[self.Start]
            while True:
                i += 1
                if currNode.GetPointer() == 0:
                    break
            return i
                
    def IsEmpty(self):
        return self.Start == 0
    
    def IsFull(self):
        return self.NextFree > 10
    
    def display(self):
        print('Start:', self.Start)
        print('NextFree:', self.NextFree)
        print('%-15s %-15s %-15s' % ('Index', 'Name', 'Pointer'))
        for i in range(1, 11):
            node = self.Node[i]
            print('%-15s %-15s %-15s' % (i, node.GetName(), node.GetPointer()))
            
    def traverse(self):
        currNode = self.Node[self.Start]
        while True:
            print(currNode.GetName(), end=',')
            if currNode.GetPointer() == 0:
                break
            currNode = self.Node[currNode.GetPointer()]
    
    
```

```python
lst = LinkedList()
lst.Insert('Ali', 1)
lst.Insert('Jack', 1)
lst.Insert('Ben',2)
lst.Delete(1)
lst.Insert('Jane', 2) 
lst.Insert('Ken', 3)   # Ben, Jane, Ken, Ali
lst.Delete(2)
lst.display()
print()
lst.traverse()
```
