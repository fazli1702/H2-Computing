```python
def read():
    print('-'*50)
    print('{:5} {:30} {}'.format('No', 'Task', 'Complete'))
    print('-'*50)

    with open('TODOLIST.txt', 'r') as f:
        data = f.readlines()
        for i, ele in enumerate(data):
            ele = ele.strip().split(',')
            task = ele[0]
            complete = ''
            if ele[1] == '1':
                complete = 'Yes'
            else:
                complete = 'No'
            print('{:<5} {:<30} {}'.format(i+1, task, complete))

    print('-'*50)
    print()
    
    
def add(new_task):
    with open('TODOLIST.txt', 'a') as f:
        f.write(new_task + ',0\n')
    print('Task added!')
    print()
    
    
def check(complete_task):
    task = []
    with open('TODOLIST.txt', 'r') as f:
        data = f.readlines()
        for ele in data:
            task.append(ele.strip().split(','))
        task[complete_task-1][1] = '1'

    with open('TODOLIST.txt', 'w') as f:
        for ele in task:
            string = ele[0] + ',' + ele[1] + '\n'
            f.write(string)

    print(f'Task {complete_task} marked as Complete')
    print()
    
    

def main():
    while True:
        print('To-Do List')
        print('1. List Tasks to do')
        print('2. Input Task to do')
        print('3. Check Task as Complete')
        print('4. Exit')
        choice = input('Choice:')
        
        if choice == '1':
            read()
                    
        elif choice == '2':
            new_task = input('Enter Task:')
            add(new_task)
        
        elif choice == '3':
            complete_task = int(input('Mark which task number as complete:'))
            check(complete_task)
        
        elif choice == '4':
            break
```

```python
main()
```
