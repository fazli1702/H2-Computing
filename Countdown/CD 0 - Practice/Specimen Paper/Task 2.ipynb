```python
# Task 2.1

def read():
    with open('TESTDATA.txt') as f:
        data = f.readlines()
        data = [row.strip().split(',') for row in data]
    return data

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    
    pivot = arr[0]
    left = []
    right = []
    
    for i in range(1, len(arr)):
        if arr[i][0] < pivot[0]:
            left.append(arr[i])
        else:
            right.append(arr[i])
    
    return quicksort(left) + [pivot] + quicksort(right)

def main():
    with open('TESTDATA_Filename_order.txt', 'w') as f:
        data = quicksort(read())
        for row in data:
            f.write(','.join(row) + '\n')
            
main()
```

```python
# Task 2.2

def read():
    with open('TESTDATA.txt') as f:
        data = f.readlines()
        data = [row.strip().split(',') for row in data]
    return data

# bubble sort
def sort_date(arr):
    end = len(arr) - 1
    for i in range(len(arr)):
        swap = False
        for j in range(end):
            if arr[j][1] > arr[j+1][1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap = True
        
        if not swap:
            break
        end -= 1
    return arr

def search_by_name(data):
    user_name = input('Enter file name: ')
    found = False
    for name, date, size in data:
        if user_name == name:
            print('-'*20)
            print('{:<10} {}'.format('file name:', name))
            print('{:<10} {}'.format('file size:', size))
            found = True
            
    if not found:
        print('File name not found')
        
        
def search_by_date(data):
    print('Date modified?')
    user_year = input('Enter year: ')
    user_month = input('Enter month: ')
    user_day = input('Enter day: ')
    user_date = user_year + '-' + user_month + '-' + user_day
    
    for name, date, size in data:
        found = False
        if user_date == date:
            print('-'*20)
            print('{:<10} {}'.format('file name:', name))
            print('{:<10} {}'.format('file date:', date))
            found = True
    
    if not found:
        print('Modified file date not found')
```

```python
data = sort_date(read())
search_by_name(data)
```
