```python
# Task 4.1
class Node:
    def __init__(self):
        self.Subject = ''
        self.LeftPtr = 0
        self.RightPtr = 0
        
    def getSubject(self):
        return self.Subject
    
    def getLeftPtr(self):
        return self.LeftPtr
    
    def getRightPtr(self):
        return self.RightPtr
    
    def setSubject(self, new):
        self.Subject = new
        
    def setLeftPtr(self, new):
        self.LeftPtr = new
        
    def setRightPtr(self, new):
        self.RightPtr = new
        
        
class Tree:
    def __init__(self):
        self.SubjectTree = [Node() if i != 0 else None for i in range(31)]
        self.Root = 0
        self.NextFree = 1
        
        for i in range(1, 30):
            self.SubjectTree[i].setLeftPtr(i+1)
            
            
    def AddSubject(self, NewItem):
        newNode = self.SubjectTree[self.NextFree]
        newNode.setSubject(NewItem)
        newNode.setLeftPtr(0)
        
        if self.Root == 0:
            self.Root = 1
            self.NextFree += 1
            
        else:
            # traverse tree
            PreviousPtr = 0
            CurrentPtr = self.Root
            LastMove = 'X'
            while CurrentPtr != 0:
                PreviousPtr = CurrentPtr
                if NewItem < self.SubjectTree[CurrentPtr].getSubject():
                    CurrentPtr = self.SubjectTree[CurrentPtr].getLeftPtr()
                    LastMove = 'Left'
                else:
                    CurrentPtr = self.SubjectTree[CurrentPtr].getRightPtr()
                    LastMove = 'Right'
                    
            if LastMove == 'Left':
                self.SubjectTree[PreviousPtr].setLeftPtr(self.NextFree)
            
            elif LastMove == 'Right':
                self.SubjectTree[PreviousPtr].setRightPtr(self.NextFree)
            
            self.NextFree += 1
            
            
    def LabelDelete(self, subject):
        if self.Root == 0:
            return 'Cannot delete from empty tree'
        
        else:
            currPtr = self.Root
            while True:
                currNode = self.SubjectTree[currPtr]
                if currNode.getSubject() == subject:
                    currNode.setSubject('deleted')
                    break
                    
                elif subject < currNode.getSubject():
                    currPtr = currNode.getLeftPtr()
                
                else:
                    currPtr = currNode.getRightPtr()
                    
    def traverse(self, pointer=1):
        if self.Root == 0:
            return 'Empty tree'
        
        else:
            currNode = self.SubjectTree[pointer]
            if currNode.getLeftPtr() != 0:
                self.traverse(currNode.getLeftPtr())
            
            print(currNode.getSubject())
            
            if currNode.getRightPtr() != 0:
                self.traverse(currNode.getRightPtr())
        
        
        
    
    def display(self):
        print('Root:', self.Root)
        print('NextFree:', self.NextFree)
        print('{:<10} {:<15} {:<10} {:<10}'.format('Index', 'Subject', 'LeftPtr', 'RightPtr'))
        for i in range(1, 31):
            node = self.SubjectTree[i]
            print('{:<10} {:<15} {:<10} {:<10}'.format(i, node.getSubject(), node.getLeftPtr(), node.getRightPtr()))
```

```python
# Task 4.4
def BuildTree():
    tree = Tree()
    with open('SUBJECT.txt') as f:
        data = f.readlines()
        subjects = [ele.strip() for ele in data]
        for subject in subjects:
            tree.AddSubject(subject)
        tree.display()
        tree.traverse()
        
BuildTree()
```
