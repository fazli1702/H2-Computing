```python
# Task 3.1

class Node:
    def __init__(self):
        self.data = ''
        self.leftPtr = -1
        self.rightPtr = -1
        
    def setData(self, s):
        self.data = s
        
    def setLeftPtr(self, x):
        self.leftPtr = x
        
    def setRightPtr(self, y):
        self.rightPtr = y
        
    def getData(self):
        return self.data
    
    def getLeftPtr(self):
        return self.leftPtr
    
    def getRightPtr(self):
        return self.rightPtr
        
class Tree:
    def __init__(self):
        self.tree = []
        self.root = -1
        
    def add(self, newItem):
        newNode = Node()
        newNode.setData(newItem)
        newIndex = len(self.tree) # index of new node
        self.tree.append(newNode)
        
        # empty tree
        if self.root == -1:
            self.root = 0
        
        else:
            # traverse tree to link child node with parent node
            currIndex = self.root
            while True:
                currNode = self.tree[currIndex]
                
                if newItem < currNode.getData():
                    if currNode.getLeftPtr() == -1:
                        currNode.setLeftPtr(newIndex)
                        break
                    else:
                        currIndex = currNode.getLeftPtr()
                
                else:
                    if currNode.getRightPtr() == -1:
                        currNode.setRightPtr(newIndex)
                        break
                    else:
                        currIndex = currNode.getRightPtr()
            
    
    def print(self):
        print('{:<15} {:<15} {:<15} {:<15}'.format('Index', 'Data', 'leftPtr', 'rightPtr'))
        for i, ele in enumerate(self.tree):
            print('{:<15} {:<15} {:<15} {:<15}'.format(i, 
            ele.getData(), ele.getLeftPtr(), ele.getRightPtr()))
```

```python
# Task 3.2
tree = Tree()
tree.add('Dave')
tree.add('Fred')
tree.add('Ed')
tree.add('Greg')
tree.add('Bob')
tree.add('Cid')
tree.add('Ali')
tree.print()
```

```python
# Task 3.3

class Node:
    def __init__(self):
        self.data = ''
        self.leftPtr = -1
        self.rightPtr = -1
        
    def setData(self, s):
        self.data = s
        
    def setLeftPtr(self, x):
        self.leftPtr = x
        
    def setRightPtr(self, y):
        self.rightPtr = y
        
    def getData(self):
        return self.data
    
    def getLeftPtr(self):
        return self.leftPtr
    
    def getRightPtr(self):
        return self.rightPtr
        
class Tree:
    def __init__(self):
        self.tree = []
        self.root = -1
        
    def add(self, newItem):
        newNode = Node()
        newNode.setData(newItem)
        newIndex = len(self.tree) # index of new node
        self.tree.append(newNode)
        
        # empty tree
        if self.root == -1:
            self.root = 0
        
        else:
            # traverse tree to link child node with parent node
            currIndex = self.root
            while True:
                currNode = self.tree[currIndex]
                
                if newItem < currNode.getData():
                    if currNode.getLeftPtr() == -1:
                        currNode.setLeftPtr(newIndex)
                        break
                    else:
                        currIndex = currNode.getLeftPtr()
                
                else:
                    if currNode.getRightPtr() == -1:
                        currNode.setRightPtr(newIndex)
                        break
                    else:
                        currIndex = currNode.getRightPtr()
            
    
    def print(self):
        print('{:<15} {:<15} {:<15} {:<15}'.format('Index', 'Data', 'leftPtr', 'rightPtr'))
        for i, ele in enumerate(self.tree):
            print('{:<15} {:<15} {:<15} {:<15}'.format(i, 
            ele.getData(), ele.getLeftPtr(), ele.getRightPtr()))
            
    def iot(self, curr):
        if curr != -1:
            currNode = self.tree[curr]
            
            # traverse left sub tree
            self.iot(currNode.getLeftPtr())
            
            print(currNode.getData())
            
            # traverse right sub tree
            self.iot(currNode.getRightPtr())
            
    def inOrderTraversal(self):
        return self.iot(self.root)
```

```python
tree = Tree()
tree.add('Dave')
tree.add('Fred')
tree.add('Ed')
tree.add('Greg')
tree.add('Bob')
tree.add('Cid')
tree.add('Ali')
tree.inOrderTraversal()
```
