```python
# Task 3.1
class ToDo:
    def __init__(self, c, d):
        self.category = c
        self.description = d
        
    def set_category(self, s):
        self.category = s
        
    def set_descritption(self, s):
        self.description = s
        
    def get_category(self):
        return self.category
    
    def get_description(self):
        return self.description
    
    def compare_with(self, td):
        if self.category < td.get_category():
            return -1
        
        elif self.category == td.get_category():
            if self.description < td.get_description():
                return -1
            elif self.description > td.get_description():
                return 1 
            else:
                return 0
        
        else:
            return 1
    
    def summary(self):
        print('category:', self.category)
        print('description:', self.description)
```

```python
# Task 3.2
def main():
    activities = [
    ToDo("reading", "Try some Shakespeare"),
    ToDo("shopping", "Consider items to recycle"),
    ToDo("reading", "Search on the web"),
    ToDo("reading", "Go to the library")
    ]

    output = []

    for activity in activities:
        if len(output) == 0:
            output.append(activity)
        else:
            insert = False
            for i, ele in enumerate(output):
                if activity.compare_with(ele) == -1:
                    output.insert(i, activity)
                    insert = True
                    break

            # end of list
            if not insert:
                output.append(activity)


    for ele in output:
        ele.summary()
        print('-'*20)

main()
```

```python
# Task 3.3
class DatedToDo(ToDo):
    def __init__(self, dt, c, d):
        super().__init__(c, d)
        self.due_date = dt
        
    def set_due_date(self, s):
        self.due_date = s
        
    def get_due_date(self):
        return self.due_date
    
    def summary(self):
        super().summary()
        print('due date:', self.due_date)
        
    def compare_with(self, td):
        if type(DatedToDo('', '', '')) == type(td):
            if self.due_date < td.get_due_date():
                return -1
            elif self.due_date > td.get_due_date():
                return 1
            else:
                super().compare_with(td)
        
        else:
            super().compare_with(td)
```
