```python
# Task 2.1

def read():
    words = []
    with open('SONG.TXT') as f:
        data = f.readlines()
        for row in data:
            row = row.strip().split(' ')
            for word in row:
                words.append(word)
    return words


def quicksort(arr):
    return quicksortHelper(arr, 0, len(arr)-1)

def quicksortHelper(arr, low, high):
    if low < high:
        mid = partition(arr, low, high)
        quicksortHelper(arr, low, mid-1)
        quicksortHelper(arr, mid+1, high)
    return arr

def partition(arr, low, high):
    pivot = arr[low]
    left = low + 1
    right = high
    done = True
    
    while done:
        while left <= right and arr[left] <= pivot:
            left += 1
            
        while left <= right and arr[right] >= pivot:
            right -= 1
            
        if left > right:
            done = False
        
        else:
            arr[left], arr[right] = arr[right], arr[left]
            
    arr[right], arr[low] = arr[low], arr[right]
    return right

def main():
    words = read() # read file and store in data structure
    quicksort(words) # sort words using quick sort
    count = {}
    for word in words:
        if word in count:
            count[word] += 1
        else:
            count[word] = 1
            
    with open('WORDCOUNT.TXT', 'w') as f:
        for word, freq in count.items():
            s = word + ' ' + str(freq)
            f.write(s + '\n')
        
main()
```

```python
# Task 2.2

def read2():
    words = []
    with open('WORDCOUNT.TXT') as f:
        data = f.readlines()
        for row in data:
            row = row.strip().split(' ')
            row[1] = int(row[1])
            words.append(row)
    return words


def binary_search(word, words):
    left = 0
    right = len(words) - 1
    found = False
    while left <= right:
        mid = (left + right) // 2
        
        if words[mid][0] == word:
            print('-'*20)
            print('Word:', words[mid][0])
            print('Count:', words[mid][1])
            print('-'*20)
            found = True
            break
        
        elif word < words[mid][0]:
            right = mid - 1
        else:
            left = mid + 1
            
    if not found:
        print('Not Found')
        
        
def search_count(count, words):
    found = False
    for word, freq in words:
        if freq == count:
            print('-'*20)
            print('Word:', word)
            print('Count:', freq)
            print('-'*20)
            found = True
            
    if not found:
        print('Not Found')
            
        
def menu():
    words = read2()
    while True:
        print('1. Search for a word using binary search algorithm')
        print('2. Search for word(s) based on count')
        print('3. Quit program')
        print()
        
        choice = input('Enter your choice: ')
        
        if choice == '1':
            word = input('Enter word: ')
            binary_search(word, words)
        
        if choice == '2':
            # validate input
            while True:
                count = input('Enter count: ')
                if count.isdigit():
                    count = int(count)
                    break
                    
            search_count(count, words)
        
        if choice == '3':
            break
            
```
