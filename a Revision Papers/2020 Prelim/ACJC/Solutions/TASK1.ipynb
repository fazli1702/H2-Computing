```python
import random

#Task 1.1
class Food():
    def __init__(self, name, value):
        self.name = name
        self.value = value

    def get_name(self):
        return self.name

    def get_value(self):
        return self.value

#Task 1.2
class Person():
    def __init__(self, name):
        self.name = name
        self.health = 100
        self.strength = 100

    def get_name(self):
        return self.name

    def get_health(self):
        return self.health

    def get_strength(self):
        return self.strength

    def eat(self, food):
        to_add = food.get_value()
        self.strength += to_add
        print("Your strength is now "+str(self.get_strength()))

    def attack(self, opponent):
        value = int(random.random()*10)+1
        if self.strength < value:
            print("You don't have enough strength to attack. Try again later.")
        else:
            opponenthealth = opponent.get_health()
            if opponenthealth < value:
                opponent.health = 0
                print(opponent.get_name()+" has been defeated.")
            else:
                opponent.health -= value
                print(opponent.get_name()+"'s strength is now "+str(opponenthealth)+".")
            self.strength -= value

#Task 1.3
class Healer(Person):
    def heal(self, other):
        value = int(random.random()*10)+1
        if self.strength < value:
            print("You don't have enough strength to heal. Try again later.")
        else:
            otherhealth = other.get_health()
            if otherhealth < 100-value:
                other.health += value
            else:
                other.health = 100
            print(other.get_name()+"'s health is now "+str(otherhealth))

#Task 1.4
class Warrior(Person):
    def attack(self, opponent):
        value = int(random.random()*10)+1
        if self.strength < value:
            print("You don't have enough strength to attack. Try again later.")
        else:
            value *= 2
            opponenthealth = opponent.get_health()
            if opponenthealth < value:
                opponent.health = 0
                print(opponent.get_name()+" has been defeated.")
            else:
                opponent.health -= value
                print(opponent.get_name()+"'s strength is now "+str(opponenthealth)+".")

#Testing
Sam = Warrior("Sam")
Alex = Healer("Alex")
Jan = Person("Jan")
Cheese = Food("Cheese", 10)

Jan.attack(Sam)
Sam.attack(Jan)
Alex.heal(Jan)
```
