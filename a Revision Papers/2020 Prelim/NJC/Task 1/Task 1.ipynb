```python
# Task 1.1
import random

def read():
    with open('WORDS.TXT') as f:
        data = f.readlines()
        words = {}
        minim = maxim = 0
        for ele in data:
            ele = ele.strip()
            minim = len(ele)  # give temp value 
            maxim = len(ele)  # give temp value
            if len(ele) not in words:
                words[len(ele)] = [ele]
            else:
                words[len(ele)] += [ele]


    for k, v in words.items():
        if k < minim:
            minim = k
        if k > maxim:
            maxim = k
            
    return maxim, minim, words
        

def getSecret():
    maxim, minim, words = read()
    user_len = int(input(f'Choose the length of word to guess:[{minim} to {maxim}]'))
    secretWord = random.choice(words[user_len])
    return secretWord

```

```python
# Task 1.2
def main():
    secretWord = getSecret()
    guessCount = len(secretWord) * 2
    
    userWord = ['.' for i in range(len(secretWord))]
    
    guess = False  # flag
    print(f'You have {guessCount} attempts. Good Luck!')
    
    for i in range(guessCount, 0, -1):
        print(''.join(userWord))
        
        userLetter = input(f'[{i}]Guess letter:')
        for i, ele in enumerate(secretWord):
            if userLetter == ele:
                userWord[i] = ele
                break
                
        if ''.join(userWord) == secretWord:
            guess = True
            break
    
    print(''.join(userWord))
    if guess:
        print('WIN')
    else:
        print('LOSE..The word is ' + secretWord)

```

```python
# Task 1.3
import random

def main():
    secretWord = getSecret()
    guessCount = len(secretWord) * 2
    n = round(0.2 * (len(secretWord)))  # number of words to add at start
    
    userWord = ['.' for i in range(len(secretWord))]
    for i in range(n):
        while True:
            k = random.randint(0, len(secretWord)-1)
            if userWord[k] == '.':
                userWord[k] = secretWord[k]
                break
        
    guess = False  # flag
    print(f'You have {guessCount} attempts. Good Luck!')
    
    for i in range(guessCount, 0, -1):
        print(''.join(userWord))
        
        userLetter = input(f'[{i}]Guess letter:')
        for i, ele in enumerate(secretWord):
            if userLetter == ele:
                userWord[i] = ele
                break
                
        if ''.join(userWord) == secretWord:
            guess = True
            break
    
    print(''.join(userWord))
    if guess:
        print('WIN')
    else:
        print('LOSE..The word is ' + secretWord)
```
