```python
# Task 2.1

import csv

def read():
    books = []
    with open('story.csv') as f:
        data = csv.reader(f)
        next(data)
        for row in data:
            books.append(row)
    return books

def MergeSort(ArrayData, SortBy):
    if len(ArrayData) <= 1:
        return ArrayData
    
    mid = len(ArrayData) // 2
    left = ArrayData[:mid]
    right = ArrayData[mid:]
    return merge(MergeSort(left, SortBy), MergeSort(right, SortBy), SortBy)

def merge(left, right, SortBy):
    lst = []
    if SortBy == 'B':
        key = 0
    if SortBy == 'S':
        key = 1
    if SortBy == 'A':
        key = 2
    if SortBy == 'P':
        key = 3
        
    while len(left) != 0 and len(right) != 0:
        if left[0][key] < right[0][key]:
            lst.append(left.pop(0))
        else:
            lst.append(right.pop(0))
            
    if len(left) != 0:
        lst += left
    if len(right) != 0:
        lst += right
        
    return lst

def main():
    ArrayData = read()  # read data from file
    
    # validate user input 
    validChoice = ['B', 'S', 'A', 'P']
    while True:
        SortBy = input('Which field do you want to sort by (B, S, A, P): ')
        if SortBy in validChoice:
            break
            
    # implement merge sort
    ArrayData = MergeSort(ArrayData, SortBy)
    for ele in ArrayData:
        print(ele)
    
    
```

```python
main()
```

```python
main()
```

```python
# Task 2.2
def QuickSort(ArrayData):
    if len(ArrayData) <= 1:
        return ArrayData
    
    left = []
    right = []
    pivot = ArrayData[0]
    for i in range(1, len(ArrayData)):
        if ArrayData[i][0] < pivot[0]:
            left.append(ArrayData[i])
        else:
            right.append(ArrayData[i])
            
    return QuickSort(left) + [pivot] + QuickSort(right)
```

```python
QuickSort(read())
```
