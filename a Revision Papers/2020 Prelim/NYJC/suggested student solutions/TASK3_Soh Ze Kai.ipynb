```python
#Task 3.1
class Node:
    def __init__(self, title: str):
        self._title = str(title)
        self._prev = None #None denotes NULL pointer
        self._next = None
        
class SongPlaylist:
    def __init__(self):
        self._head = None
        
    def insert(self, title: str):
        node = Node(str(title))
        node._next = self._head #Set next pointer for new node
        
        if self._head is not None: #Set prev pointer for previously HEAD node
            self._head._prev = node
            
        self._head = node #Set HEAD pointer
        
        
    def insertafter(self, searchtitle: str, newtitle: str):
        newnode = Node(str(newtitle))
        curr = self._head
        
        while curr is not None: #Traverse to search title
            if curr._title == searchtitle:
                #Assign pointers for next node
                nextnode = curr._next
                if nextnode is not None:
                    nextnode._prev = newnode
                
                #Assign pointer for new node and prev nodes
                newnode._next = nextnode
                curr._next = newnode
                newnode._prev = curr
                
            curr = curr._next

    def display(self):
        curr = self._head
        while curr is not None: #Traverse the linked list
            print(curr._title)
            curr = curr._next
```

```python
#Task 3.2
import sys

def menu():
    print("1. Create New Song Playlist")
    print("2. Add a song in front")
    print("3. Add a song after")
    print("4. Display Playlist")
    print("5. -1 to End")
    
playlists = {} #Dictionary to store all the playlists

while True:
    menu()
    
    option = int(input("Enter an option: "))
    
    if option == 1:
        name = input("Enter name of new playlist: ")
        playlists[name] = SongPlaylist()
        
    elif option == 2:
        title = input("Enter song title: ")
        name = input("Enter playlist name: ")
        playlists[name].insert(title)
        
    elif option == 3:
        existing_title = input("Enter existing song title: ")
        new_title = input("Enter new song title: ")
        name = input("Enter playlist name: ")
        playlists[name].insertafter(existing_title, new_title)
        
    elif option == 4:
        name = input("Enter playlist name: ")
        print("Contents of playlist:")
        playlists[name].display()
        
    elif option == -1:
        sys.exit()
        
    print()
```

```python
#Task 3.3
def insertionSort(Playlist):
    sorted = SongPlaylist() #Create temp sorted list
    sorted._head = Playlist._head
    p_curr = Playlist._head._next
    s_curr = sorted._head
    
    while p_curr is not None: #Traverse original list
        while p_curr._title > s_curr._title: #Traverse sorted list to find appropriate position
            if s_curr.next is not None:
                s_curr = s_curr._next
        
        #Insert at appropriate position
        if s_curr._prev is not None: 
            sorted.insertafter(s_curr._prev, p_curr)
        else:
            sorted.insert(p_curr)
            
        p_curr = p_curr._next
        s_curr = sorted._head
    
    Playlist._head = sorted._head #Set the HEAD pointer
    
    return Playlist
        
    
```
