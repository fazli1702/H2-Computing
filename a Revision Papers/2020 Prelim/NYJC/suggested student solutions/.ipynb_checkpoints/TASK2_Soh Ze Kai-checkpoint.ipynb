```python
#Task 2.1
def MergeSort(ArrayData, Sortby):
    return mergesort_helper(ArrayData, Sortby, 0, len(ArrayData) - 1)

def mergesort_helper(ArrayData, Sortby, left, right):
    if left < right:
        mid = (left + right) // 2
        
        #Recursively split the array into 2 halves
        mergesort_helper(ArrayData, Sortby, left, mid)
        mergesort_helper(ArrayData, Sortby, mid + 1, right)
        
        return merge(ArrayData, Sortby, left, mid, right)
    
def merge(ArrayData, Sortby, left, mid, right):
    #Copy the data into left and right temp arrays
    left_arr = ArrayData[left:mid+1]
    right_arr = ArrayData[mid+1:right+1]
    
    #Pointers for left (i), right (j) and original (k) arrays to do merging
    i = 0
    j = 0
    k = left
    
    while i < len(left_arr) and j < len(right_arr):
        if left_arr[i][Sortby] <= right_arr[j][Sortby]:
            ArrayData[k] = left_arr[i]
            i += 1
        else:
            ArrayData[k] = right_arr[j]
            j += 1
        k += 1
        
    #Copy leftover records in temp array back to original array
    while i < len(left_arr):
        ArrayData[k] = left_arr[i]
        i += 1
        k += 1
        
    while j < len(right_arr):
        ArrayData[k] = right_arr[j]
        j += 1
        k += 1
        
    return ArrayData

def main():
    records = [] #Array of records

    with open("story.csv", 'r') as infile:
        infile.readline() #Ignore headers

        for line in infile.readlines():
            line = line.rstrip('\n')
            row = line.split(',')
            
            for i in range(len(row)): #Strip any trailing or leading whitespace
                row[i] = row[i].strip()
                
            record = {'B': row[0], 'S': row[1], 'A': row[2], 'P': row[3]}
            records.append(record)

        infile.close()

    sort_by = input("Enter field to sort by (B, S, A, P): ").upper()

    while sort_by not in ["B", "S", "A", "P"]: #Validate input
        sort_by = input("Enter field to sort by (B, S, A, P): ").upper()

    records = MergeSort(records, sort_by)

    print("Contents of ArrayData")
    for record in records:
        print(record)
        
main()
```

```python
#Task 2.2
def QuickSort(ArrayData):
    quicksorthelper(ArrayData, 0, len(ArrayData) - 1)
    return ArrayData

def quicksorthelper(ArrayData, first, last):
    if first < last:
        splitpoint = partition(ArrayData, first, last)
        
        quicksorthelper(ArrayData, first, splitpoint - 1)
        quicksorthelper(ArrayData, splitpoint + 1, last)
        
    return ArrayData
    
def partition(ArrayData, first, last):
    #Lomuto Partition Scheme (single pointer implementation)
    pivot = ArrayData[last]
    pointer = first
    
    #Shift all elements more than pivot to the front
    for index in range(first+1, last):
        if ArrayData[index]["B"] > pivot["B"]:
            ArrayData[index] = temp
            ArrayData[index] = ArrayData[pointer]
            ArrayData[pointer] = temp
            pointer += 1
    
    #Assign pivot to appropriate position
    ArrayData[last] = ArrayData[pointer]
    ArrayData[pointer] = pivot
    
    return pointer
```
