```python
# Task 3.1

class Node:
    def __init__(self, title):
        self.title = title
        self.prev = None
        self.next = None
        
    def getTitle(self):
        return self.title
    
    def getPrev(self):
        return self.prev
    
    def getNext(self):
        return self.next
    
    def setPrev(self, s):
        self.prev = s
        
    def setNext(self, n):
        self.next = n
        

class SongPlaylist:
    def __init__(self):
        self.head = None
        
    def insert(self, title):
        newNode = Node(title)
        newNode.setNext(self.head)
        self.head = newNode
    
    def insertafter(self, searchtitle, newtitle):
        # before: [prevNode] <-> [currNode]
        # after : [prevNode] <-> [newNode] <-> [currNode]
        newNode = Node(newtitle)
        prevNode = None  
        currNode = self.head
        
        # traverse linked list
        while True:
            if currNode.getTitle() == searchtitle:
                prevNode = currNode
                currNode = currNode.getNext()
                break
            else:
                prevNode = currNode
                currNode = currNode.getNext()
                
        if currNode != None:
            prevNode.setNext(newNode)        
            newNode.setPrev(prevNode)
            newNode.setNext(currNode)
            currNode.setPrev(newNode)
        else:
            prevNode.setNext(newNode)
            newNode.setPrev(prevNode)
        
    
    def display(self):
        if self.head == None:
            print('Empty song playlist')
        else:
            currNode = self.head
            while currNode != None:
                print(currNode.getTitle())
                currNode = currNode.getNext()
```
