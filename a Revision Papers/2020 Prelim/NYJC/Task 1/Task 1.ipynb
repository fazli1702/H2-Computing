```python
# Task 1.1

def read():
    lst = []
    with open('waste.csv') as f:
        data = f.readlines()
        for i, row in enumerate(data):
            row = row.strip()
            if i == 0:
                continue
            else:
                row = [ele for ele in row.strip().split(',')]
                row[0], row[1], row[2] = int(row[0]), float(row[1]), float(row[2])
                lst.append(row)
    return lst

def getKeyAddress(Year):
    return Year % 20
            
def initTable():
    return ['' for i in range(20)]

def insert(data, table):
    year, dispose, recycle = data
    i = getKeyAddress(year)
    if table[i] == '':
        table[i] = data
    else:
        while table[i] != '':
            i = (i+1) % 20
        table[i] = data
        
def main():
    data = read()
    table = initTable()
    for ele in data:
        insert(ele, table)
    return table

main()
```

```python
# Task 1.2

def getInfo(year, table):
    i = getKeyAddress(year)
    if table[i] != '':
        if table[i][0] == year:
            return table[i]
    else:
        while table[i][0] != '':
            i = (i+1) % 20
        return table[i]

def recycleGreaterDisposed():
    years = []
    data = read()
    for year, recycle, dispose in data:
        if recycle > dispose:
            years.append(year)
            
    return years

def averageDispose(year1, year2, table):
    average = []
    for i in range(year1, year2+1):
        info = getInfo(i, table)
        average.append(info[1])
    average = sum(average) / len(average)
    return average


def menu():
    table = main()
    while True:
        print()
        print('1. Get Waste Disposed and Recycled by year')
        print('2. Display year(s) where Recycled waste > Waste disposed')
        print('3. Return Average waste disposed between two years')
        print('4. -1 to Exit')
        print()
        
        user_choice = input('Enter your choice:')
        
        if user_choice == '1':
            while True:
                user_year = input('Enter year:')
                if user_year.isdigit():
                    if 2000 <= int(user_year) <= 2014:
                        user_year = int(user_year)
                        break
                        
            info = getInfo(user_year, table)
            print(f'Year: {user_year}')
            print(f'Waste Disposed: {info[1]} (tonnes)')
            print(f'Waste Recycled: {info[2]} (tonnes)')
            
        
        elif user_choice == '2':
            years = recycleGreaterDisposed()
            print(years)
        
        elif user_choice == '3':
            while True:
                user_year1 = input('Enter start year:')
                if user_year1.isdigit():
                    if 2000 <= int(user_year1) <= 2014:
                        user_year = int(user_year1)
                        break
                        
            while True:
                user_year2 = input('Enter end year:')
                if user_year2.isdigit():
                    if 2000 <= int(user_year2) <= 2014:
                        user_year2 = int(user_year2)
                        break
                        
            average = averageDispose(int(user_year1), int(user_year2), table)
            print(f'Average between {user_year1} and {user_year2}: {average}')
        
        elif user_choice == '4':
            break
 

menu()
```
