```python
# Task 2.1

class Queue:
    def __init__(self):
        self.Items = ['' for i in range(5)]
        self.Front = -1
        self.Rear = -1
        
    def IsEmpty(self):
        return self.Front == -1
    
    def IsFull(self):
        return self.Rear == 4
    
    def Enqueue(self, s):
        if self.Rear == 4:
            return 'Queue is full'
        
        if self.Front == -1:
            self.Front = 0
            self.Rear = 0
            self.Items[self.Rear] = s
        else:
            self.Rear += 1
            self.Items[self.Rear] = s
    
    def Dequeue(self):
        if self.IsEmpty():
            return None
        else:
            temp = self.Items[self.Front]
            self.Front += 1
            return temp
    
    def Display(self):
        lst = []
        for i in range(self.Front, self.Rear+1):
            lst.append(self.Items[i])
        print(lst)
        
    def info(self):
        print('front:', self.Front)
        print('rear:', self.Rear)
        print('items:', self.Items)
        
            
class CircularQueue(Queue):
    def __init__(self):
        super().__init__()
        
    def IsFull(self):
        i = (self.Rear + 1) % 5
        return i == self.Front
            
    def Enqueue(self, s):
        if self.IsFull():
            return 'Queue is full'
        
        # empty queue
        elif self.Front == -1:
            self.Front = 0
            self.Rear = 0
            self.Items[self.Front] = s
            
        else:
            self.Rear = (self.Rear + 1) % 5
            self.Items[self.Rear] = s
    
    def Dequeue(self):
        if self.IsEmpty():
            return None
        else:
            temp = self.Items[self.Front]
            self.Front = (self.Front + 1) % 5
            return temp
    
    def Display(self):
        if self.Front == -1:
            return 'Empty Queue'
        
        currIndex = self.Front
        lst = []
        while True:
            currItem = self.Items[currIndex]
            lst.append(currItem)
            
            if currIndex == self.Rear:
                break
            else:
                currIndex = (currIndex + 1) % 5
                
        print(lst)
        
    def info(self):
        print('front:', self.Front)
        print('rear:', self.Rear)
        print('items:', self.Items)
```

```python
# Task 2.2

def main():
    q = Queue()
    cq = CircularQueue()
    names = ['John', 'Amy', 'Chetan','Xin Xin', 'Evan']
    for name in names:
        q.Enqueue(name)
        cq.Enqueue(name)
        
    for i in range(2):
        q.Dequeue()
        cq.Dequeue()
        
    q.Enqueue('mohan')
    cq.Enqueue('mohan')
    
    print('Queue:')
    q.Display()
    print()
    
    print('Circular Queue:')
#     cq.info()
    cq.Display()
    
    
main()
```
