```python
class BSTNode:
    def __init__(self):
        self._ISBN = ''
        self._Title = ''
        self._Left = None
        self._Right = None
        
    def GetISBN(self):
        return self._ISBN
    
    def SetISBN(self, s):
        self._ISBN = s
        
    def GetTitle(self):
        return self._Title
    
    def SetTitle(self, s):
        self._Title = s
        
    def GetLeft(self):
        return self._Left
    
    def SetLeft(self, s):
        self._Left = s
        
    def GetRight(self):
        return self._Right
    
    def SetRight(self, s):
        self._Right = s
        
    def Display(self):
        print(self._ISBN, self._Title)
        

class BST:
    def __init__(self):
        self._Root = None
        
    def GetRoot(self):
        return self._Root
    
    def SetRoot(self, s):
        self._Root = s
        
    def AddNode(self, isbn, title):
        newNode = BSTNode()
        newNode.SetISBN(isbn)
        newNode.SetTitle(title)
        
        if self.GetRoot() == None:
            self.SetRoot(newNode)
        else:
            currNode = self.GetRoot()
            while True:
                if isbn < currNode.GetISBN():
                    if currNode.GetLeft() == None:
                        currNode.SetLeft(newNode)
                        break
                    else:
                        currNode = currNode.GetLeft()
                
                else:
                    if currNode.GetRight() == None:
                        currNode.SetRight(newNode)
                        break
                    else:
                        currNode = currNode.GetRight()
            
    
    def _Display(self, curr):
        # traverse left sub tree
        if curr.GetLeft() != None:
            self._Display(curr.GetLeft())

        print(curr.GetTitle())
        
        # traverse right sub tree
        if curr.GetRight():
            self._Display(curr.GetRight())
        
    def Display(self):
        return self._Display(self._Root)
            
                
```

```python
# Task 3.2
def CalcHash(my_string):
    total = 0
    for i in range(len(my_string)):
        total += ord(my_string[i]) * (i+1)
    
    total %= 19
    total += 1
    return total
```

```python
# Task 3.3
def init_table():
    table = ['' for i in range(19)]
    return table
```

```python
# Task 3.4

def read():
    with open('CATEGORIES.TXT') as f:
        data = f.readlines()
        categories = [ele.strip() for ele in data]
    return categories


def insert(ele, table):
    i = CalcHash(ele)
    if table[i] == '':
        table[i] = [ele, BST()]
    else:
        while True:
            if table[i] == '':
                break
            else:
                i = (i+1) % 19
        table[i] = [ele, BST()]
        
        
def main():
    categories = read()
    HashTable = init_table()
    for category in categories:
        insert(category, HashTable)
    return HashTable

main()
```

```python
# Task 3.5
def read2():
    with open('BOOKS.TXT') as f:
        data = f.readlines()
        books = [ele.strip().split(',') for ele in data]
    return books

def main2():
    books = read2()
    HashTable = main()
    for category, isbn, title in books:
        i = CalcHash(category)
        if HashTable[i][0] == category:
            tree = HashTable[i][1]
        else:
            while HashTable[i][0] != category:
                i = (i+1) % 19
            tree = HashTable[i][1]

        tree.AddNode(isbn, title)
        
    return HashTable
```
