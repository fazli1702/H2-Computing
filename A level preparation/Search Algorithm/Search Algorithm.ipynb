```python
Value = (2, 6, 15, 23, 36, 48, 50, 58, 64, 69, 74, 79, 86, 92, 99) # do not modify

def binary_search(score):
    low = 0
    high = len(Value) - 1
    
    search = []
    while not low > high:
        middle = (low + high) // 2
        search.append(Value[middle])
        
        if Value[middle] == score:
            return middle, search
        
        elif Value[middle] < score:
            low = middle + 1
        
        elif Value[middle] > score:
            high = middle - 1
            
    return None, search
    
```

```python
print(binary_search(64))
print(binary_search(80))
```

```python
a=\
''''
Get SearchName
Set HashTotal to 0
FOR each Character in SearchName
    Get the ASCII code for Character
    Multiply the ASCII code by the position of Character in SearchName
    Add the result to HashTotal
Calculate Hash as HashTotal MOD 500
RETURN HASH
'''
```

```python
def GenerateHash(SearchName):
    HashTotal = 0
    i = 1
    for Character in SearchName:
        HashTotal += ord(Character) * i
        i += 1
    Hash = HashTotal % 500
    return Hash
```
