```python
class Food:
    def __init__(self, n, c, d):
        self.name = n
        self.category = c
        self.description = d
        
    def get_name(self):
        return self.name
    
    def get_category(self):
        return self.category
    
    def get_description(self):
        return self.description
    
    def set_name(self, new):
        self.name = new
        
    def set_category(self, new):
        self.category = new
        
    def set_description(self, new):
        self.description = new
        
    def summary(self):
        print('Name:', self.name)
        print('Category:', self.category)
        print('Description:', self.description)
        
    def compare_with(self, food):
        # compare category
        if self.category < food.get_category():
            return -1
        
        # within category
        elif self.category == food.get_category():
            if self.name < food.get_name():
                return -1
            elif self.name == food.get_name():
                return 0
            else:
                return 1
        
        else:
            return 1
```

```python
# DON'T HAVE TO READ FROM TEXT FILE
# quetion did not say read from text file
# just copy past from text file
food_lst = [Food("Apple", "Fruits", "6 Sweet Pink, bought on 23 May"),
Food("Milk", "Diary", "Low fat, 1 litre"),
Food("Carrot", "Vegetables", "medium sized, Australian"),
Food("Chicken meat", "Protein", "thigh meat, 200g"),
Food("Cabbage", "Vegetables", "300g")]
```

```python
lst = []
for food in food_lst:
    # empty list
    if lst == []:
        lst.append(food)
        
    else:
        add = True
        for i in range(len(lst)):
            add = False
            if food.compare_with(lst[i]) == -1:
                lst.insert(i, food)
                add = True
                break
        
        # end of lst
        if not add:
            lst.append(food)
            
                
```

```python
for food in lst:
    print(food.summary())
    print('-----------------')
```

```python
class DatedFood(Food):
    def __init__(self, n, c, d, dt):
        super().__init__(n, c, d)
        self.expiry_date = dt
        
    def set_expiry_date(self, new):
        self.expiry_date = new
        
    def get_expiry_date(self):
        return self.expiry_date
    
    def compare_with(self, food):
        if self.expiry_date < food.get_expiry_date():
            return -1
        elif self.expiry_date == food.get_expiry_date():
            super().compare_with(food)
        else:
            return 1
    
    def summary(self):
        super().summary()
        print('Expiry Date:', self.expiry_date)
```
