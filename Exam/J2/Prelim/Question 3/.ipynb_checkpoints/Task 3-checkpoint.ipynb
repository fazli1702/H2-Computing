```python
# Task 3.3
class Node:
    def __init__(self):
        self.word = ''
        self.category = ''
        self.next = -1
        
    def set_word(self, w):
        self.word = w
        
    def set_category(self, c):
        self.category = c
        
    def set_next(self, n):
        self.next = n
        
    def get_word(self):
        return self.word
    
    def get_category(self):
        return self.category
    
    def get_next(self):
        return self.next
    

class LinkedList:
    def __init__(self):
        self.sentence = [Node() for i in range(10)]
        self.start = -1
        self.nextfree = 0
        for i in range(9):
            self.sentence[i].set_next(i+1)
        
    def isempty(self):
        return self.start == -1
    
    def isfull(self):
        return self.nextfree == -1
    
    def display(self):
        print('start:', self.start)
        print('nextfree:', self.nextfree)
        
        print('{:<10} {:<10} {:<10} {:<10}'.format('index', 'word', 'category', 'next'))
        for i in range(10):
            currNode = self.sentence[i]
            print('{:<10} {:<10} {:<10} {:<10}'.format(i, currNode.get_word(), currNode.get_category(), currNode.get_next()))
    
    def insert(self, new_word, new_category):
        if self.nextfree == -1:
            return 'Linked List is Full'
        
        self.sentence[self.nextfree].set_word(new_word)
        self.sentence[self.nextfree].set_category(new_category)
        
        # insert into empty tree
        if self.start == -1:
            self.start = self.nextfree
            temp = self.sentence[self.start].get_next()
            self.sentence[self.start].set_next(-1)
            self.nextfree = temp
            
        # traverse linked list & search for position to insert
        else:
            current = self.start
            previous = -1
            inserted = False  # flag
            
            while current > -1 and not inserted:
                if self.sentence[current].get_category() > new_category:
                    if current == self.start:
                        self.start = self.nextfree
                    else:
                        self.sentence[previous].set_next(self.nextfree)
                        
                    temp = self.sentence[self.nextfree].get_next()
                    self.sentence[self.nextfree].set_next(current)
                    self.nextfree = temp
                    inserted = True
                    
                elif self.sentence[current].get_category() < new_category:
                    previous = current
                    current = self.sentence[current].get_next()
                    
                else:
                    if new_category == 'N':
                        new_category = 'Z'
                    
                    if new_category == 'D':
                        new_category = 'X'
                        
                    if new_category == 'J':
                        new_category = 'Y'
                        
                    previous = current
                    current = self.sentence[current].get_next()
                    self.sentence[self.nextfree].set_category(new_category)
                    
            if not inserted:
                self.sentence[previous].set_next(self.nextfree)
                temp = self.sentence[self.nextfree].get_next()
                self.sentence[self.nextfree].set_next(-1)
                self.nextfree = temp
            
        
    
    def traversal(self):
        pass
```

```python
# Task 3.5
class Node:
    def __init__(self):
        self.word = ''
        self.category = ''
        self.next = -1
        
    def set_word(self, w):
        self.word = w
        
    def set_category(self, c):
        self.category = c
        
    def set_next(self, n):
        self.next = n
        
    def get_word(self):
        return self.word
    
    def get_category(self):
        return self.category
    
    def get_next(self):
        return self.next
    

class LinkedList:
    def __init__(self):
        self.sentence = [Node() for i in range(10)]
        self.start = -1
        self.nextfree = 0
        for i in range(9):
            self.sentence[i].set_next(i+1)
        
    def isempty(self):
        return self.start == -1
    
    def isfull(self):
        return self.nextfree == -1
    
    def display(self):
        print('start:', self.start)
        print('nextfree:', self.nextfree)
        
        print('{:<10} {:<10} {:<10} {:<10}'.format('index', 'word', 'category', 'next'))
        for i in range(10):
            currNode = self.sentence[i]
            print('{:<10} {:<10} {:<10} {:<10}'.format(i, currNode.get_word(), currNode.get_category(), currNode.get_next()))
    
    def insert(self, new_word, new_category):
        if self.nextfree == -1:
            return 'Linked List is Full'
        
        self.sentence[self.nextfree].set_word(new_word)
        self.sentence[self.nextfree].set_category(new_category)
        
        # insert into empty tree
        if self.start == -1:
            self.start = self.nextfree
            temp = self.sentence[self.start].get_next()
            self.sentence[self.start].set_next(-1)
            self.nextfree = temp
            
        # traverse linked list & search for position to insert
        else:
            current = self.start
            previous = -1
            inserted = False  # flag
            
            while current > -1 and not inserted:
                if self.sentence[current].get_category() > new_category:
                    if current == self.start:
                        self.start = self.nextfree
                    else:
                        self.sentence[previous].set_next(self.nextfree)
                        
                    temp = self.sentence[self.nextfree].get_next()
                    self.sentence[self.nextfree].set_next(current)
                    self.nextfree = temp
                    inserted = True
                    
                elif self.sentence[current].get_category() < new_category:
                    previous = current
                    current = self.sentence[current].get_next()
                    
                else:
                    if new_category == 'N':
                        new_category = 'Z'
                    
                    if new_category == 'D':
                        new_category = 'X'
                        
                    if new_category == 'J':
                        new_category = 'Y'
                        
                    previous = current
                    current = self.sentence[current].get_next()
                    self.sentence[self.nextfree].set_category(new_category)
                    
            if not inserted:
                self.sentence[previous].set_next(self.nextfree)
                temp = self.sentence[self.nextfree].get_next()
                self.sentence[self.nextfree].set_next(-1)
                self.nextfree = temp
            
        
    
    def traversal(self):
        if self.start == -1:
            return 'Empty linked list'
        
        else:
            string = ''
            currIndex = self.start
            while True:
                currNode = self.sentence[currIndex]
                if currNode.get_category() == 'X':
                    currNode.set_category('D')
                    
                if currNode.get_category() == 'Y':
                    currNode.set_category('J')
                    
                if currNode.get_category() == 'Z':
                    currNode.set_category('N')
                    
                string += currNode.get_word() + ' '
                
                if currNode.get_next() == -1:
                    break
                    
                else:
                    currIndex = currNode.get_next()
                    
        print(string)
```

```python
lst = LinkedList()
lst.insert('little', 'J')
lst.insert('tough', 'J')
lst.insert('boy', 'N')
lst.insert('The', 'D')
lst.insert('runs', 'V')
lst.insert('race', 'N')
lst.insert('a', 'D')
lst.traversal()
```
