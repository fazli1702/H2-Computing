```python
# Task 2.1

def quick_sort(arr):
    quick_sort_helper(arr, 0, len(arr)-1)
    return arr

def quick_sort_helper(arr, low, high):
    if low < high:
        mid = partition(arr, low, high)
        quick_sort_helper(arr, low, mid-1)
        quick_sort_helper(arr, mid+1, high)
    return arr

def partition(arr, low, high):
    pivot = arr[low]
    left = low + 1
    right = high
    done = True
    
    while done:
        while left <= right and arr[left] <= pivot:
            left += 1
            
        while left <= right and arr[right] >= pivot:
            right -= 1
            
        if left > right:
            done = False
            
        else:
            arr[left], arr[right] = arr[right], arr[left]
            
    arr[low], arr[right] = arr[right], arr[low]
    return right



def read():
    words = []
    with open('SONG.TXT') as f:
        data = f.readlines()
        for row in data:
            row = row.strip().split(' ')
            words += row
    
    quick_sort(words)
    return words
    
    
def main():
    words = read()
    freq = {}
    for word in words:
        if word not in freq:
            freq[word] = 1
        else:
            freq[word] += 1
    
    with open('WORDCOUNT.TXT', 'w') as f:
        for word, n in freq.items():
            string = word + ' ' +  str(n) + '\n'
            f.write(string)
    

main()
```

```python
# Task 2.2
def read2():
    words = []
    count = []
    with open('WORDCOUNT.TXT') as f:
        data = f.readlines()
        for row in data:
            row = row.strip().split(' ')
            words.append(row)        
    return words


def binary_search(word, arr):
    high = len(arr)-1
    low = 0
    
    while low <= high:
        mid = (high+low) // 2
        
        if arr[mid][0] == word:
            return arr[mid]
        
        elif arr[mid][0] > word:
            high = mid - 1
            
        else:
            low = mid + 1
            
    return 'Not Found'



def main2():
    words = read2()
    while True:
        print('1. Search for word')
        print('2. Search for word based on count')
        print('3. Quit program')
        
        choice = input('Enter your choice:')
        
        if choice == '1':
            user_word = input('Enter your word:')
            output = binary_search(user_word, words)
            print(output)
        
        elif choice == '2':
            user_choice = input('Enter count:')
            for word in words:
                if word[1] == user_choice:
                    print(word)
        
        elif choice == '3':
            break
            
        print()
```
